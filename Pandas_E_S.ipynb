### normalizando archivos json

```python
import json

# Cargar el archivo JSON
with open("/content/drive/MyDrive/Panda E_S/pacientes_2.json", 'r', encoding='utf-8') as f:
    data = json.load(f)

# Normalizar los datos, extrayendo la lista de pacientes como registros
df = pd.json_normalize(data, record_path='Pacientes', meta=['Investigación', 'Año'], max_level=1)

# Expandir la columna 'Problemas_salud' en columnas separadas
problemas_df = df['Problemas_salud'].apply(pd.Series)

# Renombrar las columnas (opcional)
problemas_df.columns = [f'Problema_salud_{i+1}' for i in problemas_df.columns]

# Unir las columnas expandidas al DataFrame original (y eliminar la original si se desea)
df = pd.concat([df.drop(columns=['Problemas_salud']), problemas_df], axis=1)

# Mostrar el resultado
df
```
